```python
# Function to clean text
def clean_text(text):
    # Remove special characters (like !, , .) and make lowercase
    text = re.sub(r"[^\w\s]", "", text).lower()
    # Remove stopwords
    words = text.split()
    cleaned_words = [word for word in words if word not in stopwords]
    return " ".join(cleaned_words)

cleaned_docs = {key: clean_text(value) for key, value in docs.items()}
```

# Apply lemmatization

```python
import re
import nltk
from nltk.corpus import wordnet
from nltk.stem import WordNetLemmatizer
from nltk import pos_tag, download

# Download required resources (only once)
download('punkt')
nltk.download('averaged_perceptron_tagger_eng')
download('wordnet')
download('omw-1.4')

lemmatizer = WordNetLemmatizer()

# Helper to convert POS tag to WordNet tag
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN

# Lemmatize each document
lemmatized_docs = {}
for key, sentence in cleaned_docs.items():
    words = sentence.split()
    pos_tags = pos_tag(words)
    lemmatized_words = [
        lemmatizer.lemmatize(word, get_wordnet_pos(tag))
        for word, tag in pos_tags
    ]
    lemmatized_docs[key] = " ".join(lemmatized_words)
```

```python
from collections import defaultdict
import math
# Step 1: Build vocabulary
vocab = sorted(set(word for doc in lemmatized_docs.values() for word in doc.split()))

vocab
```

# TF Function (term frequency)

```python
def compute_tf(doc):
    tf = defaultdict(float)
    words = doc.split()
    for word in words:
        tf[word] += 1
    total_words = len(words)
    for word in tf:
        tf[word] /= total_words
    return tf
```

# IDF Function (inverse document frequency)

```python
def compute_idf(docs):
    N = len(docs)
    idf = defaultdict(float)
    for word in vocab:
        containing_docs = sum(1 for doc in docs.values() if word in doc.split())
        idf[word] = math.log((N + 1) / (containing_docs + 1)) + 1  # smooth IDF
    return idf
```

# TF-IDF Function

```python
def compute_tfidf(tf, idf):
    return [tf[word] * idf[word] for word in vocab]
```

```python
# Compute IDF once for all docs
idf = compute_idf(lemmatized_docs)

# Compute TF-IDF for each doc
tfidf_vectors = {}
for doc_id, text in lemmatized_docs.items():
    tf = compute_tf(text)
    tfidf = compute_tfidf(tf, idf)
    tfidf_vectors[doc_id] = tfidf

# Show results
for doc_id, vector in tfidf_vectors.items():
    print(f"TF-IDF for doc {doc_id}: {vector}")
```

# Calculate Prior Probabilities 𝑃(𝑦)

```python
def count(number , l):
  c=0
  for i in l:
    if i == number:
      c+=1
  return c


for i in list(set(y)):
  print(f"P({i}) = {count(i,y)/len(y)}")
```

```python
def calculate_prior_probabilities(y):
    total_docs = len(y)
    class_counts = defaultdict(int)
    for label in y:
        class_counts[label] += 1
    priors = {label: count / total_docs for label, count in class_counts.items()}
    return priors

priors = calculate_prior_probabilities(y)
print("Prior Probabilities P(y):", priors)
```

# Calculate Likelihood  𝑃(𝑥∣𝑦) with Laplace Smoothing

$$ P(x \mid y) = \frac{\text{count of word in class } y + 1}{\text{total words in class } y + |V|} $$

```python
from collections import defaultdict

def calculate_likelihoods(X, y, vocab):
    class_word_counts = defaultdict(lambda: defaultdict(int))  # Word counts for each class
    class_word_totals = defaultdict(int)  # Total word count for each class

    # Count the word occurrences per class
    for doc, label in zip(X, y):
        for word_index, word_value in enumerate(doc):
            if word_value > 0:  # Only consider non-zero values (non-zero words in the vector)
                word = vocab[word_index]
                class_word_counts[label][word] += 1
                class_word_totals[label] += 1

    # Apply Laplace Smoothing and calculate probabilities
    likelihoods = defaultdict(lambda: defaultdict(float))
    vocab_size = len(vocab)

    for label in class_word_counts:
        for word in class_word_counts[label]:
            # Laplace smoothing: (count + 1) / (total words in class + vocab size)
            likelihoods[label][word] = (class_word_counts[label][word] + 1) / (class_word_totals[label] + vocab_size)

    return likelihoods, class_word_totals


likelihoods, class_word_totals = calculate_likelihoods(X, y, vocab)

# Print the likelihoods in a more readable format
print("Likelihoods P(x|y) with Laplace Smoothing:")
for label, word_probs in likelihoods.items():
    print(f"Class {label}:")
    for word, prob in word_probs.items():
        print(f"  P({word}|{label}) = {prob:.4f}")
```

# Calculate Posterior Probabilities for Each Class

```python
import math

def calculate_posterior_probabilities(X, y, priors, likelihoods, vocab):
    posterior_probs = []

    for doc in X:
        log_prob_class_1 = math.log(priors.get(1, 1e-6))  # Log probability to avoid underflow
        log_prob_class_0 = math.log(priors.get(0, 1e-6))  # Log probability to avoid underflow

        # For each word in the document
        for word_index, word_value in enumerate(doc):
            if word_value > 0:  # Only consider non-zero values (non-zero words in the vector)
                word = vocab[word_index]

                # For class 1
                log_prob_class_1 += math.log(likelihoods[1].get(word, 1e-6))  # Apply smoothing

                # For class 0
                log_prob_class_0 += math.log(likelihoods[0].get(word, 1e-6))  # Apply smoothing

        # Store posterior probabilities for the current document
        posterior_probs.append((log_prob_class_1, log_prob_class_0))

    return posterior_probs
```

# Naive Bayes for Classification

$$ P(y \mid x) \propto P(y) \prod_{i=1}^{n} P(x_i \mid y) $$

```python
# Compute posterior probabilities for the training documents
posterior_probs = calculate_posterior_probabilities(X, y, priors, likelihoods, vocab)

# Print the posterior probabilities
for i, (log_prob_1, log_prob_0) in enumerate(posterior_probs):
    print(f"Document {i+1}: P(1|x) = {math.exp(log_prob_1):.6f}, P(0|x) = {math.exp(log_prob_0):.6f}")
```

# Make Predictions Based on Posterior Probabilities

```python
# Prediction Based on Posterior Probabilities
def predict(posterior_probs):
    predictions = []
    for log_prob_1, log_prob_0 in posterior_probs:
        if log_prob_1 > log_prob_0:
            predictions.append(1)  # Spam (Class 1)
        else:
            predictions.append(0)  # Not Spam (Class 0)
    return predictions

# Make predictions for the documents
predictions = predict(posterior_probs)

# Print predictions
for i, pred in enumerate(predictions):
    print(f"Document {i+1}: Predicted class = {pred}")
```

# . Evaluate Accuracy (Optional)

```python
def evaluate(predictions, y_true):
    correct = sum(p == t for p, t in zip(predictions, y_true))
    accuracy = correct / len(y_true)
    return accuracy

# Evaluate the model
accuracy = evaluate(predictions, y)
print(f"Accuracy: {accuracy:.2f}")
```

# Handling Unknown Words (UNK)

```python
import math
from collections import defaultdict

# 1. Helper: compute posterior for a single TF‑IDF vector // helper that takes a single TF‑IDF vector and returns its posterior per class
def posterior_for_tfidf(tfidf, priors, likelihoods, vocab, smooth=1e-10):
    post = {}
    for label, prior in priors.items():
        logp = math.log(prior)
        for i, val in enumerate(tfidf):
            if val > 0:
                w = vocab[i]
                # if word never seen in this class, fall back to smooth
                like = likelihoods[label].get(w, smooth)
                logp += math.log(like)
        post[label] = logp
    return post

# 2. Updated predict (only passes what the helper needs)
def predict_new_sentence_with_UNK(sentence, vocab, priors, likelihoods, idf):
    # map unknown words → "UNK"
    words = [w if w in vocab else "UNK" for w in sentence.split()]
    # recompute TF over that list
    tf = defaultdict(float)
    for w in words: tf[w] += 1
    for w in tf: tf[w] /= len(words)
    # build TF‑IDF vector in same vocab order
    tfidf = [tf[w] * idf[w] for w in vocab]
    # get posterior log‑probs
    post = posterior_for_tfidf(tfidf, priors, likelihoods, vocab)
    # choose the class with the max log‑posterior
    return max(post, key=post.get)
```

# Ignore words approach

```python
def preprocess_ignore_unknown(sentence, vocab):
    # نحتفظ بس بالكلمات اللي موجودة في vocab
    return [w for w in sentence.split() if w in vocab]
import math

def posterior_single(tfidf, priors, likelihoods, vocab, smooth=1e-10):
    logp = {}
    for label, prior in priors.items():
        lp = math.log(prior)
        for i,val in enumerate(tfidf):
            if val > 0:
                w = vocab[i]
                # لو الكلمة ما ظهرتش في likelihoods[label] نستخدم قيمة تنعيم صغيرة
                lp += math.log(likelihoods[label].get(w, smooth))
        logp[label] = lp
    return logp

from collections import defaultdict

def predict_ignore_unknown(sentence, vocab, priors, likelihoods, idf):
    # 1. نحتفظ بالكلمات المعروفة فقط
    words = preprocess_ignore_unknown(sentence, vocab)
    # 2. نحسب TF بناءً على الكلمات المتبقية
    tf = defaultdict(float)
    for w in words: tf[w] += 1
    for w in tf: tf[w] /= len(words) if words else 1

    # 3. نحسب TF-IDF بنفس دوال التدريب
    tfidf = [tf.get(w, 0) * idf.get(w, 0) for w in vocab]

    # 4. نحسب الاحتمالات البوستيريور
    post = posterior_single(tfidf, priors, likelihoods, vocab)

    # 5. نرجّع الفئة الأعلى احتمال
    return max(post, key=post.get)
```

```python
new_sentence = "buy a cheap new watch"
predicted = predict_ignore_unknown(new_sentence, vocab, priors, likelihoods, idf)
print(f"Predicted class: {predicted}")
```

# Vocabulary Expansion

```python
from nltk.corpus import wordnet

def expand_vocabulary(vocab):
    expanded_vocab = set(vocab)
    for word in vocab:
        synonyms = set()
        for syn in wordnet.synsets(word):
            for l in syn.lemmas():
                synonym = l.name().replace("_", " ").lower()
                if synonym != word:
                    synonyms.add(synonym)
        expanded_vocab.update(synonyms)
    return sorted(list(expanded_vocab))

expanded_vocab = expand_vocabulary(vocab)
print("Expanded Vocabulary:", expanded_vocab)


# Recompute TF-IDF with expanded vocabulary

def compute_idf_expanded(docs, expanded_vocab):
    N = len(docs)
    idf = defaultdict(float)
    for word in expanded_vocab:
        containing_docs = sum(1 for doc in docs.values() if word in doc.split())
        idf[word] = math.log((N + 1) / (containing_docs + 1)) + 1  # smooth IDF
    return idf

idf_expanded = compute_idf_expanded(lemmatized_docs, expanded_vocab)

#Compute TF-IDF for each doc with expanded vocab
tfidf_vectors_expanded = {}
for doc_id, text in lemmatized_docs.items():
    tf = compute_tf(text)
    tfidf = [tf.get(word,0) * idf_expanded.get(word,0) for word in expanded_vocab]
    tfidf_vectors_expanded[doc_id] = tfidf

# Show results
for doc_id, vector in tfidf_vectors_expanded.items():
    print(f"TF-IDF for doc {doc_id}: {vector}")

X_expanded = list(tfidf_vectors_expanded.values())

likelihoods_expanded, class_word_totals_expanded = calculate_likelihoods(X_expanded, y, expanded_vocab)

# Print the likelihoods in a more readable format
print("Likelihoods P(x|y) with Laplace Smoothing:")
for label, word_probs in likelihoods_expanded.items():
    print(f"Class {label}:")
    for word, prob in word_probs.items():
        print(f"  P({word}|{label}) = {prob:.4f}")


```
